<h1 style = "color: #454644">Data Cleaning</h1>

```python
# some columns won't be useful in getting insights from the data neither the modeling phase.
# as well as they have many null values so I will drop them.
useless_columns = ["Prize", "Motivation", "Full Name", "Laureate ID"]
df.drop(useless_columns, axis = 1, inplace = True)
```

<h3 style = "color: #454644">Category</h3>

```python
df["Category"].value_counts()
```

That's good we don't have any strange values.

```python
df["Category"] = df["Category"].astype("category")
```

<h3 style = "color: #454644">Prize share</h3>

```python
df["Prize Share"].value_counts()
```

Prize share value indicates how many people shared the prize together.
So We can make sure that the data in it is correct by comparing it with year and category columns duplicates.

```python
df["Prize Share"] = df["Prize Share"].astype("category")
```

<h3 style = "color: #454644">Laureate Type</h3>

```python
df["Laureate Type"].value_counts()
```

```python
df["Laureate Type"] = df["Laureate Type"].astype("category")
```

<h3 style = "color: #454644">Birth columns and Sex</h3>

```python
Birth_cols = ["Birth Date", "Birth Country", "Birth City", "Sex"]
```

```python
df["Sex"].value_counts()
```

```python
# we need to convert Birth data column to datatime
df["Birth Date"] = pd.to_datetime(df["Birth Date"], errors = "coerce")
```

```python
df["Birth Date"].fillna(df["Birth Date"].median(), inplace = True)
for col in  Birth_cols[1:]:
    df[col].fillna("unknown", inplace = True)
    df[col] = df[col].astype("category")
```

```python
# I will add a column that has the age of the winner when he won the prize
df["Win Age"] = df["Year"] - df["Birth Date"].dt.year
```

```python
df.info()
```

<h3 style = "color: #454644">Organization columns</h3>

```python
org_cols = ["Organization Name", "Organization City", "Organization Country"]
```

```python
for col in org_cols:
    df[col].fillna("unknown", inplace = True)
    df[col] = df[col].astype("category")
```

<h3 style = "color: #454644">Death Columns</h3>

```python
death_cols = ["Death Date", "Death City", "Death Country"]
```

```python
df["Death Date"] = pd.to_datetime(df["Death Date"], errors = "coerce")
df["Death Date"].fillna(datetime.datetime.now(), inplace = True)
```

```python
for col in death_cols[1:]:
    df[col].fillna("unknown", inplace = True)
    df[col] = df[col].astype("category")
```

<h1 style = "color: #454644">EDA</h1>

```python
def count(data, x, title, xlabel , order = None ,hue = None, annotate = False, xticks = None):
    g = sns.countplot(data = data, x = x, order = order, palette = "pastel", hue = hue)
    if annotate:
        for p in g.patches:
            g.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                        ha='center', va='bottom',
                        color= 'black', weight = "bold", fontsize = 11)
    g.set_title(title, weight = "bold", fontsize = 20, fontname = "monospace")
    g.set_xlabel(xlabel, weight = "semibold", fontsize = 15, fontname = "monospace")
    g.set_ylabel("Count", weight = "semibold", fontsize = 15, fontname = "monospace")
    if xticks:
        g.set_xticklabels(xticks)
    plt.show()
```

```python
x = df["Year"].value_counts().sort_index().index
y = df["Year"].value_counts().sort_index().values
g = sns.lineplot(x = x, y = y)
g.set_title("Number of Prizes", weight = "bold", fontsize = 25)
g.set_xlabel("Year", weight = "semibold", fontsize = 15)
g.set_ylabel("Count", weight = "semibold", fontsize = 15)
plt.show()
```

```python
order = df["Year"].value_counts().sort_values(ascending = False)[:15].index
title = "Number of Prizes for each year"
xlabel = "Year"
count(df, "Year", title, xlabel , order = order, annotate = True)
```
